```python
f=[]

for i in range(pdfReader.getNumPages()):
    pageObj = pdfReader.getPage(i)
    #d.append({'Page_Num': i+1, 'Content': sent_tokenize(pageObj.extractText())})
    
    #d = [y for x in d for y in x] 
    z=(nltk.sent_tokenize(pageObj.extractText()))
    r_nwnt=re.compile("[\n\t\r]")
    z2 = [(r_nwnt.sub("",s)) for i,s in enumerate(z)]
    
    #print(z3)
    for z1 in z2 :

        f.append({'Page_Num': i+1, 'Content': z1}) 
df6 = pd.DataFrame(f)
```
